```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 2], [2, 2], [2, 3],[8, 7], [8, 8], [25, 80]])
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
db = DBSCAN(eps = 1, min_samples= 2)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps = 1, min_samples= 3)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps = 1, min_samples= 5)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps = 5, min_samples= 2)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps = 9, min_samples= 2)
db.fit(X)
db.labels_
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np


X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)
db = DBSCAN(eps = 0.1, min_samples = 5)
model = db.fit_predict(X)
```

```python
X[:5]
```

```python
model
```

```python
plt.scatter(X[:, 0], X[:, 1], c= model, cmap='bwr', marker='o')
plt.title("DBSCAN")
plt.xlabel("X[0]")
plt.ylabel("X[1]")
plt.show()
```
