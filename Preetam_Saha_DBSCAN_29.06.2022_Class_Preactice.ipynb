# spiral

```python
sns.scatterplot(x=df.iloc[:,0],y=df.iloc[:,1],c=df.iloc[:,2])
```

```python
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import StandardScaler
```

```python
scaled_data = StandardScaler().fit_transform(df.iloc[:,:2])
scaled_data[:5]
```

```python
dist_mat = pairwise_distances(scaled_data)
```

```python
dist_mat
```

```python
dist_mat.shape
```

```python
import matplotlib.pyplot as plt
```

```python
dist_mat.flatten()
```

```python
len(dist_mat.flatten())
```

```python
312*312
```

```python
h=plt.hist(dist_mat.flatten(),bins=50)
```

```python
plt.figure(figsize=(10,8))
sns.histplot(dist_mat.flatten())
```

```python
plt.figure(figsize=(10,8))
sns.distplot(dist_mat.flatten())
```

```python
from scipy.stats import gaussian_kde
```

```python
import numpy as np
```

```python
kde=gaussian_kde(dist_mat.flatten())
kde
```

```python
input_array = np.linspace(start=0,stop=0.3,num=300)
input_array
```

```python
density_vals = kde(input_array)
density_vals
```

```python
sns.histplot(density_vals)
```

```python
plt.hist(density_vals)
```

```python
eps = input_array[density_vals==max(density_vals)]
eps
```

```python
pts = (dist_mat<=eps).sum()
```

```python
pts
```

```python
dist_mat<=eps
```

```python
pts = (dist_mat<=eps).sum(axis=0)
```

```python
pts
```

```python
h = plt.hist(pts,bins=20)
```

```python
min_pts = 2
```

```python
from sklearn.cluster import DBSCAN, OPTICS
```

```python
dbs = DBSCAN(eps=eps[0],min_samples=min_pts)
dbs
```

```python
dbs_cluster = dbs.fit_predict(scaled_data)
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=dbs_cluster)
```

# concentric circles

```python
from sklearn.datasets import make_circles
```

```python
df1, target = make_circles(n_samples=1000,noise=0.05,factor = 0.5)
```

```python
df1
```

```python
plt.scatter(df1[:,0],df1[:,1])
```

```python
dist_mat1 = pairwise_distances(df1)
```

```python
dist_mat1
```

```python
len(dist_mat1)
```

```python
h1 = plt.hist(dist_mat1.flatten(),bins=30)
```

```python
plt.figure(figsize=(15,8))
sns.distplot(dist_mat1.flatten())
```

```python
kde1 = gaussian_kde(dist_mat1.flatten())
```

```python
input_array1 = np.linspace(start=0.1,stop=0.3,num=500)
```

```python
input_array1
```

```python
density_vals1 = kde1(input_array1)
density_vals1
```

```python
eps1 = input_array1[density_vals1==max(density_vals1)]
eps1
```

```python
pts1 = (dist_mat1<=eps1).sum(axis=0)
```

```python
pts1
```

```python
plt.figure(figsize=(15,8))
h1 = plt.hist(pts1, bins=20)
```

```python
min_pts1 = 10
```

```python
from sklearn.cluster import DBSCAN, OPTICS
```

```python
dbs1 = DBSCAN(eps=eps1[0],min_samples=min_pts1)
```

```python
dbs_cluster1 = dbs1.fit_predict(df1)
```

```python
plt.scatter(df1[:,0],df1[:,1],c=dbs_cluster1)
```
