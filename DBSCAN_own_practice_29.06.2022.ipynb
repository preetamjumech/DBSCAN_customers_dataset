```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
x1,label = make_moons(n_samples=200)
x1[:5]
```

```python
x,label = make_moons(n_samples=200,noise=0.1)
x[:5]
```

```python
label
```

```python
model = DBSCAN(eps=0.25, min_samples=10)
model.fit(x)
```

```python
model.labels_
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.scatter(x[:,0],x[:,1],c=model.labels_)
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.scatter(x[:,0],x[:,1],c=label)
```

```python
model1 = DBSCAN(eps=0.2, min_samples=10)
model1.fit(x)
fig, ax = plt.subplots(figsize=(15,8))
ax.scatter(x[:,0],x[:,1],c=model1.labels_)
```

```python
model2 = DBSCAN(eps=0.3, min_samples=10)
model2.fit(x)
fig, ax = plt.subplots(figsize=(15,8))
ax.scatter(x[:,0],x[:,1],c=model2.labels_)
```
